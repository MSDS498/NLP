```python
import nltk
nltk.download('stopwords')
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
```

```python
#Tokenize
tokenizer=RegexpTokenizer(r'\w+')
SupplierA=SupplierA.apply(lambda x: tokenizer.tokenize(x.lower()))
```

```python
#Take out stopwords in Portuguese
def remove_stopwords(text):
    words=[w for w in text if w not in stopwords.words('Portuguese')]
    return words
SupplierA=SupplierA.apply(lambda x: remove_stopwords(x))
```

```python
SupplierA
```

```python
#Run Topic Clustering with LDA

#Convert to list
SupplierAReviews=[]
SupplierAReviews.append(SupplierA)
```

```python
SupplierAReviews
```

```python
#Text to Dictionary
from gensim import corpora, models
dictionary = corpora.Dictionary(SupplierA)
print(dictionary)
```

```python
#count the # of occurences  of each distinct word and convert word to integer with doc2bow (like tf-idf)
corpus=[dictionary.doc2bow(text) for text in SupplierA]
```

```python
corpus
```

```python
#LDA Modeling (Topic Modeling)
from gensim.models import LdaModel
NUM_TOPICS=10
ldamodel=LdaModel(corpus,num_topics=NUM_TOPICS,id2word=dictionary,passes=50)
```

```python
topics=ldamodel.show_topics()
for topic in topics:
    print(topics)
```

```python
word_dict = {};
for i in range (NUM_TOPICS):
    words=ldamodel.show_topic(i,topn=10)
    word_dict['Topic #'+ '{:02d}'.format(i+1)]=[i[0] for i in words]
top10=pd.DataFrame(word_dict)
top10
```

```python
#Supplier A top 10 topics extract as excel to translate to english

top10.to_excel(r'/Users/neeramajumdar/Desktop/Capstone/Top 10/SupplierA.xlsx')
```
