```python
#Text to Dictionary for each Supplier
from gensim import corpora, models
dictionarySupplierA = corpora.Dictionary(SupplierA)
dictionarySupplierB = corpora.Dictionary(SupplierB)
dictionarySupplierC = corpora.Dictionary(SupplierC)
dictionarySupplierD = corpora.Dictionary(SupplierD)
dictionarySupplierE = corpora.Dictionary(SupplierE)
dictionarySupplierF = corpora.Dictionary(SupplierF)
dictionarySupplierG = corpora.Dictionary(SupplierG)
dictionarySupplierH = corpora.Dictionary(SupplierH)
dictionarySupplierI = corpora.Dictionary(SupplierI)
dictionarySupplierJ = corpora.Dictionary(SupplierJ)
print(dictionarySupplierB)
```

```python
#count the # of occurences  of each distinct word and convert word to integer with doc2bow (like tf-idf)
corpusSupplierA=[dictionary.doc2bow(text) for text in SupplierA]
corpusSupplierB=[dictionary.doc2bow(text) for text in SupplierB]
corpusSupplierC=[dictionary.doc2bow(text) for text in SupplierC]
corpusSupplierD=[dictionary.doc2bow(text) for text in SupplierD]
corpusSupplierE=[dictionary.doc2bow(text) for text in SupplierE]
corpusSupplierF=[dictionary.doc2bow(text) for text in SupplierF]
corpusSupplierG=[dictionary.doc2bow(text) for text in SupplierG]
corpusSupplierH=[dictionary.doc2bow(text) for text in SupplierH]
corpusSupplierI=[dictionary.doc2bow(text) for text in SupplierI]
corpusSupplierJ=[dictionary.doc2bow(text) for text in SupplierJ]
```

```python
#LDA Modeling (Topic Modeling)
from gensim.models import LdaModel
NUM_TOPICS=10
ldamodelSupplierA=LdaModel(corpusSupplierA,num_topics=NUM_TOPICS,id2word=dictionary,passes=50)
ldamodelSupplierB=LdaModel(corpusSupplierB,num_topics=NUM_TOPICS,id2word=dictionary,passes=50)
ldamodelSupplierC=LdaModel(corpusSupplierC,num_topics=NUM_TOPICS,id2word=dictionary,passes=50)
ldamodelSupplierD=LdaModel(corpusSupplierD,num_topics=NUM_TOPICS,id2word=dictionary,passes=50)
ldamodelSupplierE=LdaModel(corpusSupplierE,num_topics=NUM_TOPICS,id2word=dictionary,passes=50)
ldamodelSupplierF=LdaModel(corpusSupplierF,num_topics=NUM_TOPICS,id2word=dictionary,passes=50)
ldamodelSupplierG=LdaModel(corpusSupplierG,num_topics=NUM_TOPICS,id2word=dictionary,passes=50)
ldamodelSupplierH=LdaModel(corpusSupplierH,num_topics=NUM_TOPICS,id2word=dictionary,passes=50)
ldamodelSupplierI=LdaModel(corpusSupplierI,num_topics=NUM_TOPICS,id2word=dictionary,passes=50)
ldamodelSupplierJ=LdaModel(corpusSupplierJ,num_topics=NUM_TOPICS,id2word=dictionary,passes=50)
```
